## 6. adding and removing data

```python
import pandas as pd

df = pd.read_csv(
    'data/earthquakes.csv',
    usecols=['time', 'title', 'place', 'magType', 'mag', 'alert', 'tsunami']
)
```

```python
df['source'] = 'USGS API'
df.head()
```

```python
df['mag_negative'] = df.mag < 0 # 새로운 컬럼 생성
df.head()
```

```python
df.place.str.extract(r', (.*$)')[0].sort_values().unique()
# NV, Nevada 는 같음 , CA, california도
```

```python
df['parsed_place'] = df.place.str.replace(
    r'.* of ', '', regex=True # remove anything saying <something> of <something>
).str.replace(
    'the ', '' # remove "the "
).str.replace(
    r'CA$', 'California', regex=True # fix California
).str.replace(
    r'NV$', 'Nevada', regex=True # fix Nevada
).str.replace(
    r'MX$', 'Mexico', regex=True # fix Mexico
).str.replace(
    r' region$', '', regex=True # chop off endings with " region"
).str.replace(
    'northern ', '' # remove "northern "
).str.replace(
    'Fiji Islands', 'Fiji' # line up the Fiji places
).str.replace(
    r'^.*, ', '', regex=True # remove anything else extraneous from the beginning
).str.strip() # remove any extra spaces
```

```python
df.assign( # 새로운 열 만들기
    in_ca=df.parsed_place.str.endswith('California'),
    in_alaska=df.parsed_place.str.endswith('Alaska')
).sample(5, random_state=0)
```

```python
df.assign(
    in_ca=df.parsed_place == 'California',
    in_alaska=df.parsed_place == 'Alaska',
    neither=lambda x: ~x.in_ca & ~x.in_alaska
).sample(5, random_state=0)
```

```python
tsunami = df[df.tsunami == 1]
no_tsunami = df[df.tsunami == 0]

tsunami.shape, no_tsunami.shape # df 나누기
```

```python
pd.concat([tsunami, no_tsunami]).shape # df 합치기
```

```python
del df['source']
df.columns # source 컬럼 제거
```

```python
mag_negative = df.pop('mag_negative')
df.columns
```

```python
df.drop([0, 1]).head(2) # 0,1 행 제거 axis=0 행제거가 디폴트임
```

```python
regex = r"aa"
matches = re.findall(regex, test_str) # 리스트로 반환
print(matches)
```

```python
matches = re.finditer(regex, test_str, re.MULTILINE) # 오브젝트로 반환
matches
```

1. . any character except newline
2. \w \d \s word, digit, whitespace
3. \W \D \S not word, not digit, not whitespace
4. [abc]any of a, b, or c
5. [^abc]not a, b, or c
6. [a-h] character between a & h

1. ^abc$start / end of the string
2. \b \B word boundary, not-word boundary

1. a* a+ a? 0 or more, 1 or more, 0 or 1
2. a{5} a{2,}exactly five, two or more
